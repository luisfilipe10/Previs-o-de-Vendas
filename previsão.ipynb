```python
import pandas as pd
import plotly.express as px
from sklearn.linear_model import LinearRegression
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
dados = pd.read_csv('/content/drive/MyDrive/historico.csv')
```

```python
print(dados)
```

```python
# Definição de colunas para previsão    y = valor a ser previsto (ex: vendas)
#lista_colunas = dados[colunas]         x = dados usados para prever (ex: mes)
y = dados['Vendas']
lista_colunas = ['Mês']
x = dados[lista_colunas]
```

```python
modelo = LinearRegression()
modelo.fit(x,y)
```

```python
meses_prever = pd.DataFrame({'Mês': [25,26,27,28,29,30]})
previsao = modelo.predict(meses_prever)
```

```python
print(previsao)
```

```python
tabela_previsao =  pd.DataFrame({'Mês': [25,26,27,28,29,30], 'Vendas': previsao})
print(tabela_previsao)
```

```python
dados['Tipo'] = 'Real'
tabela_previsao['Tipo'] = 'Previsto'
```

```python
tbl_grafico = pd.concat([dados, tabela_previsao])
grafico = px.scatter(tbl_grafico, x='Mês', y='Vendas', color='Tipo',trendline='ols')
grafico.show()
```
